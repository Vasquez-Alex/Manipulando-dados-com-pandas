```python
import pandas as pd

dados = {
    'Produtos': ['A', 'B', 'C', 'D'],
    'Valor': [50, 120, 180, 250],
    'Quantidade': [10, 100, 1000, 500]
}

df  = pd.DataFrame(dados)

filtro = df[df['Valor'] > 150]
print("/n os valores maiores que 150 s√£o:" )
filtro
```

```python
filtro3 = df[(df['Valor'] > 150) & (df['Quantidade'] > 10)]
print("\n Linhas com valor > 150 e Quantidade > 10 s√£o:")
filtro3
```

```python
filtro2 = df[df['Produtos'].isin(['A', 'B'])]
print("/n os  valores que tem A ou B s√£o:" )
filtro2
```
