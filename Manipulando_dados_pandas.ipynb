```python
import pandas as pd

dados = {
    'Produtos': ['A', 'B', 'C', 'D'],
    'Valor': [50, 120, 180, 250],
    'Quantidade': [10, 100, 1000, 500]
}

df  = pd.DataFrame(dados)

filtro = df[df['Valor'] > 150]
print("/n os valores maiores que 150 são:" )
filtro
```

```python
filtro2 = df[df['Produtos'].isin(['A', 'C'])]
print("\n Os produtos A e C são:")
filtro2
```

```python
filtro3 = df[(df['Valor'] > 150) & (df['Quantidade'] > 10)]
print("\n Linhas com valor > 150 e Quantidade > 10 são:")
filtro3
```

```python
filtro = df[df['Valor'].between(100, 150)]
print("\n Linhas com valor entre 100 e 150 são:")
filtro
```

```python
filtro = df[df['Produtos'].str.contains('^A')]
print("\n Linhas que contém a letra A no inicio são:")
filtro
```

```python
import pandas as pd

dados = {
    'Nome': ['Perfume', 'Shapoo', 'Esmalte', 'Desodorante'],
    'Estoque': [5, 20, 30, 50]
}

df = pd.DataFrame(dados)

filtro = df[df['Estoque'] < 50]
print("\n Os produtos que ten menos de 50 unidades são:")
filtro
```

```python
import pandas as pd

dados = {
    'Nome': ['Geraldo', 'Melissa', 'Caue', 'Dori'],
    'Matricula': ['120122DGF', '1254JLM', '457YHG', '8756DY'],
    'Departamento': ['Vendas', 'RH', 'Vendas', 'Financeiro']
}

df = pd.DataFrame(dados)

filtro = df[df['Departamento'] == 'Vendas']
print("\n Os funcionarios que trabalham no departamento de vendas são:")
filtro
```

```python
import pandas as pd

dados = {
    'Nome': ['Ryan', 'Alex', 'Quézya', 'Germano'],
    'Nota': [10, 7, 9, 1],
    'Idade': [23, 21, 19, 29]
}

df = pd.DataFrame(dados)

filtro = df[df['Idade'] > 20]
print("\n Os alunos que tem mais de 20 anos são:")
filtro
```

```python
import pandas as pd

dados = {
    'Medicamento': ['Dipirona', 'Benegripe', 'Dorflex', 'Laxante'],
    'Valor': [15, 29, 28, 8],
    'Quantidade': [39, 7, 16, 5]
}

df = pd.DataFrame(dados)

filtro = df[(df['Quantidade'] < 10) & (df['Valor'] > 20)]
print("\n Os medicamentos que tem menos de 10 unidades e custam mais de 20 reais são:")
filtro
```
